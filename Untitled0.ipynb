```python
# Step 1: Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Create dataset
data = {
    'Sleep_Hours': [4,5,6,7,8,6,5,4,7,8],
    'Study_Hours': [8,7,6,5,4,6,7,8,5,4],
    'Mobile_Usage': [6,5,4,3,2,4,5,6,3,2],
    'Caffeine_Intake': [4,3,2,2,1,3,4,5,2,1],
    'Mock_Test_Marks': [45,50,60,70,80,65,55,40,75,85],
    'Stress_Level': ['High','High','Medium','Low','Low','Medium','High','High','Low','Low']
}

df = pd.DataFrame(data)
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='Stress_Level', data=df)
plt.title("Stress Level Distribution of Students")
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Stress_Level text ni number lo marchadam
le = LabelEncoder()
df['Stress_Level'] = le.fit_transform(df['Stress_Level'])

# Inputs (X) and Output (y)
X = df.drop('Stress_Level', axis=1)
y = df['Stress_Level']

# Data split (training + testing)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Model create cheyyadam
model = LogisticRegression()
model.fit(X_train, y_train)

# Prediction
y_pred = model.predict(X_test)

print("Model Accuracy:", accuracy_score(y_test, y_pred))
```

```python
# New student details
# Sleep, Study, Mobile, Caffeine, Mock Marks
new_student = [[5, 7, 5, 3, 55]]

prediction = model.predict(new_student)

print("Predicted Stress Level:",
      le.inverse_transform(prediction))
```
